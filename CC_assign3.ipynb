```python
# Q6
import pandas as pd
employees = {
    'Employee_ID' : [101,102,103,104,105] ,
    'Name' : ["Alice","Bob","Charlie","Diana","Edward"] ,
    'Department' : ["HR","IT","IT","Marketing","Sales"] ,
    'Age' : [29,34,41,28,38] ,
    'Salary' : [50000,70000,65000,55000,60000] ,
    'Years_of_Experience' : [4,8,10,3,12] ,
    'Joining_Date' : ["2020-03-15","2017-07-19","2013-06-01","2021-02-10","2010-11-25"] ,
    'Gender' : ["Female","Male","Male","Female","Male"] ,
    'Bonus' : [5000,7000,6000,4500,5000] ,
    'Rating' : [4.5,4.0,3.8,4.7,3.5]
};

df = pd.DataFrame(employees)
print(df)
print()

print("Shape of data frame : ",df.shape)
print()
print("Data frame info : ",df.info())
print()
print("Descriptive status : ",df.describe())
print("First 5 rows : ",df.head())
print("Last 3 rows : ",df.tail(3))

average_salary = df["Salary"].mean()
total_bonus = df["Bonus"].sum()
youngest_age = df["Age"].min()
highest_rating = df["Rating"].max()

print("Average salary : ",average_salary)
print("Total bonus : ",total_bonus)
print("Youngest_employee : ",youngest_age)
print("Highest performance : ",highest_rating)

sort_salary = df.sort_values(by="Salary",ascending = False)
print("Sorted dataframe : ",sort_salary)

print()
print()

def categorize_performace(rating) :
  if rating >= 4.5 :
    return "Excellent"
  elif rating >= 4.0 :
    return "Good"
  else :
    return "Average"

df["Performace Category"] = df["Rating"].apply(categorize_performace)
print("Df with new col : ")
print(df)

print("Missing values : ")
print(df.isnull().sum())

df.rename(columns = {"Employee_ID" : "ID"},inplace = True)
print("After renaming : ")
print(df.head())

print()
experienced_employees = df[df["Years_of_Experience"]>5]
it_employees = df[df["Department"]=="IT"]

print("Employees with greater than 5 years of experience : ")
print(experienced_employees)
print("It employees : ")
print(it_employees)

print()

df["Tax"] = df["Salary"] * 0.10
print("Df with tax col : ")
print(df)

print()

df.to_csv("employees.csv",index = False)
```
