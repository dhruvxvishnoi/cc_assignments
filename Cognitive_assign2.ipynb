```python
# Q1
l = [10,20,30,40,50,60,70,80]
# add 200 and 300 to list
print("Printing list after adding : ")
l.append(200)
l.append(300)
print(l)
print()

# remove 10 and 30
l.remove(10)
l.remove(30)
print(f"List after removing : {l}")
print()

# sort in ascending
l_ascending = sorted(l)
print(f"After sorting in ascending order : {l_ascending}")
print()
# descending
l_descending = sorted(l,reverse = True)
print(f"After sorting in descending order : {l_descending}")

# Q2
print()
scores = (45,89.5,76,45.4,89,92,58,45)
highest = max(scores)
highest_index = scores.index(highest)

print(f"Highest score : {highest}")
print(f"Highest score index : {highest_index}")
print()

lowest = min(scores)
lowest_count = scores.count(lowest)

print(f"Lowest score : {lowest}")
print(f"Lowest score count : {lowest_count}")
print()

# reversing the tuple and returning it as a list because tuples are immutable and lists are mutable
reversed_scores = list(reversed(scores))
print(f"Reversed scores : {reversed_scores}")
print()

user_input = int(input("Enter a score to check : "))
if user_input in scores :
  print(f"{user_input} is present in scores")
  print(f"Its first ocurence index : {scores.index(user_input)}")
else :
  print(f"{user_input} is not present in scores")

# Q3
import random
random_numbers = [random.randint(100,900) for _ in range(100)]
print("Printing odd numbers")
for number in random_numbers :
  if(number%2!=0) :
    print(number)

print()
print("Priting even numbers")
for number in random_numbers :
  if(number%2==0) :
    print(number)

print()
print("Printing prime numbers")
def is_prime(number) :
  if number < 2 :
    return false
    for i in range(2,int(num**0.5)+1) :
      if number % i == 0 :
        return False
    return True

for number in random_numbers :
  if is_prime(number) :
    print(number)

# Q4
print()
a = {34,56,78,90}
b = {78,45,90,23}

union = a.union(b)
print(f"Printing union : {union}")
intersection = a.intersection(b)
print(f"Printing intersection : {intersection}")
symm_diff = a.symmetric_difference(b)
print(f"Printing symmetric difference : {symm_diff}")
a_subset_b = a.issubset(b)
print(f"Checking if a is subset of b : {a_subset_b}")
b_superset_a = b.issuperset(a)
print(f"Checking if b is superset of a : {b_superset_a}")

user_input = int(input("Entet a element in a : "))
if user_input in a :
  a.remove(user_input)
  print(f"{user_input} removed successfully from a")
else :
  print(f"{user_input} does not exist in a")

# Q5
print()
sample_dict = {
    "name" : "Kelly" ,
    "age" : 25 ,
    "salary" : 8000 ,
    "city" : "New York"
};

print("Original dictionary : ",sample_dict)
sample_dict["location"] = sample_dict.pop("city")
# The pop() method is used to remove the key city from the dictionary and retrieve its value.
print("Updated dict : ",sample_dict)

```
